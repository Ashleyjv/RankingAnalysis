```python
import requests
from bs4 import BeautifulSoup

def scrape_websites(urls):
    # This function takes a list of URLs and scrapes the textual content
    results = {}
    for url in urls:
        try:
            # Send a GET request to the URL
            response = requests.get(url)
            # Check if the request was successful
            if response.status_code == 200:
                # Parse the HTML content of the page
                soup = BeautifulSoup(response.text, 'html.parser')
                # Extract textual content and use strip to clean up leading/trailing whitespace
                text = soup.get_text(separator=' ', strip=True)
                # Store the extracted text of each URL in a dictionary
                results[url] = text
            else:
                # If the request failed, store the status code for the URL
                results[url] = f"Failed to retrieve. Status code: {response.status_code}"
        except Exception as e:
            # If an error occurs during the request, store the error message
            results[url] = str(e)
    return results

# Example list of URLs to scrape
urls = ["https://en.wikipedia.org/wiki/Artificial_intelligence", "https://www.oracle.com/artificial-intelligence/what-is-ai/"]
# Scrape the websites and get the results
scraped_data = scrape_websites(urls)
for url, content in scraped_data.items():
    print(f"Text content from {url}:")
    print(content[:10])  # Print only the first 500 characters to avoid too much output
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def fetch_content(url, max_length=1000):
    """ Fetches the content of a given URL with a timeout and limits the content to max_length characters """
    try:
        # Include a timeout parameter to avoid hanging on a single request
        response = requests.get(url, timeout=10)  # Timeout set to 10 seconds
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            text = soup.get_text(separator=' ', strip=True)
            return text[:max_length]
    except requests.exceptions.Timeout:
        return "Timeout: Request exceeded time limit."
    except requests.exceptions.RequestException as e:
        # Handle other requests exceptions
        return f"Request Exception: {str(e)}"
    except Exception as e:
        return f"Error: {str(e)}"
    return "Failed to retrieve content or invalid URL"

def process_csv(file_path, max_length=5000):
    # Load the CSV file, but only the first 400 rows
    df = pd.read_csv(file_path, nrows=1000)

    # Check if the 'link' column exists
    if 'link' not in df.columns:
        return "The CSV file does not have a 'link' column."

    # Initialize an empty list to store content
    contents = []
    
    # Loop over each URL in the 'link' column and fetch the content
    for i, url in enumerate(df['link']):
        content = fetch_content(url, max_length)
        contents.append(content)
        # Print the progress after each URL is processed
        print(f"Processed {i + 1}/{len(df['link'])} URLs")

    # Assign the collected content to the 'content' column
    df['content'] = contents

    # Save the modified DataFrame back to a CSV file
    df.to_csv('updated_file.csv', index=False)
    return "Updated file has been saved as 'updated_file.csv'"

# Path to your CSV file
file_path = 'cars_forsale_price_us_uk.csv'
result = process_csv(file_path)
print(result)
```
